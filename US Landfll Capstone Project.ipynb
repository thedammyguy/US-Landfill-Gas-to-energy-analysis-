```python
import matplotlib.pyplot as plt

# Getting the % distribution of operational status
status_counts = df2_cleaned['current_project_status'].value_counts(normalize=True) * 100

# Grouping smaller categories into "Others" (e.g., those < 2%)
threshold = 2
small_cats = status_counts[status_counts < threshold]
if not small_cats.empty:
    status_counts = status_counts[status_counts >= threshold]
    status_counts['Others'] = small_cats.sum()

# Define a custom color list alternating red and orange
custom_colors = ['#D62828', '#F77F00', '#003049'] * (len(status_counts) // 2 + 1)
custom_colors = custom_colors[:len(status_counts)]  # Trim to exact number of categories

# Plotting a pie chart
plt.figure(figsize=(10, 8))
status_counts.plot.pie(
    autopct='%1.1f%%',
    startangle=140,
    shadow=False,
    legend=False,
    colors=custom_colors
)

plt.title('Project Distribution by Operational Status', fontsize=16)
plt.ylabel('')  # Hide y-axis label
plt.tight_layout()
plt.show()
```

```python
# Grouping emissions reduction by state
emissions_reduced_by_states = df2_cleaned.groupby('state')['current_year_emission_reductions_(mmtco2e/yr)_-_direct'].sum().sort_values(ascending=False).head(10)

norm = plt.Normalize(emissions_reduced_by_states.min(), emissions_reduced_by_states.max())
colors = plt.cm.Greens(norm(emissions_reduced_by_states)) 

#Plotting the chart
plt.figure(figsize=(14, 6))  # Wider figure for better spacing
emissions_reduced_by_states.plot(kind='bar', color=colors)

# Labels and title
plt.title('Estimated Annual Emission reduction by State', fontsize=14)
plt.xlabel('State', fontsize=12)
plt.ylabel('Estimated Annual Emission Reduction (tons CO₂)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.show()
```
